```python
dfm['win_by']=np.where(dfm['win_by_runs']>0,'Bat first','Bowl first')
dfm.head()
```

```python
Win=dfm.win_by.value_counts()
labels=np.array(Win.index)
sizes = Win.values
colors = ['#FFFF00', '#FF0000']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=20)
plt.axis('equal',fontsize=10)
plt.show()
```

```python
plt.figure(figsize = (18,10))
sns.countplot('season',hue='win_by',data=dfm,palette='pastel')
plt.title("Matches won by batting and bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
Toss=dfm.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['#FFFF00', '#FF0000']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss result',fontsize=20)
plt.axis('equal',fontsize=10)
plt.show()
```

```python
wicketdata=dfs.dropna(subset=['dismissal_kind'])
wicketdata=wicketdata[~wicketdata['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]
```

```python
wicketdata.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True).iloc[:10,:]
```

```python
seasondata=dfm[['id','season','winner']]

completedata=dfs.merge(seasondata,how='inner',left_on='match_id',right_on='id')
```

```python
six_data=completedata[completedata['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

```python
six_data.max()
```

```python
four_data=completedata[completedata['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
four_data.max()
```
